```python
columns_to_review = ['Ventas_autos_nuevos', #La venta de autos efectivamente cayó en Abril en pandemia
 'Generacion_de_energia_electrica_CDEC_GWh', #corregir
 'Ocupacion_en_Enseanza_INE', #corregir
 'Ocupacion_en_Actividades_profesionales_INE', #corregir
 'Precio_del_propano_centavos_de_dolargalon_DTN', #corregir
 'Precio_del_petroleo_WTI_dolaresbarril', #se ve una caída fuerte pero no mayor al 50% en 1 periodo, corregir
 'Precio_del_petroleo_Brent_dolaresbarril', #igual a la anterior corregir
 'Precio_del_gas_natural_dolaresmillon_de_unidades_termicas_britanicas', #corregir
 'Precio_del_diesel_centavos_de_dolargalon', #corregir
 'Ocupacion_en_Actividades_de_organizaciones_y_organos_extraterritoriales_INE'] #reemplazar 0 con nulos y corregir
columns_to_review.insert(0, 'Periodo')

macro[columns_to_review]
```

```python
# Voy a calcular el orden de magnitud de cada celda, y compararlo con su sucesor 
# (asumo que el primer valor siempre está bien)
# Utilizando algoritmo en base 10, cualquier variación mayor a 0.5 (una caída mayor al 50% en 1 mes <ni BTC cayó tanto>)
# voy a asumir a que se trata de un error en la digitación y voy a multiplicarlo por 10
# luego esto iterará nuevamente hasta que no hayan más cambios #flexing recursive functions

def magnitude_order(serie):
    '''Pass a serie to correct the magnitude order for some entries'''
    non_shifted = macro[serie].apply(lambda x: math.log(x, 10))
    shifted = macro[serie].shift(1).apply(lambda x: math.log(x, 10))
    
    magnitude = (shifted - non_shifted) > 0.5    
    macro.loc[magnitude, serie] = macro.loc[magnitude, serie] * 10
    
    if magnitude.sum() > 0:
        magnitude_order(serie)
```

```python
#primero corrijo los valores de 'Ocupacion_en_Actividades_de_organizaciones_y_organos_extraterritoriales_INE'
col_replace_zero = 'Ocupacion_en_Actividades_de_organizaciones_y_organos_extraterritoriales_INE'
macro[col_replace_zero] = macro[col_replace_zero].replace(0, np.nan)
```

```python
#ahora, voy a correr la correción para todas las columnas excepto Periodo
columns_to_correct = macro.columns[1:]
for column in columns_to_correct:
    magnitude_order(column)
```

```python
def prep_historical_region(region: str, start_date, end_date):
    """ 
    Function to plot historic precipitation for a region. Requires region name, start and end date
    List of available regions: 
        Coquimbo, Valparaiso, Metropolitana_de_Santiago, 
        Libertador_Gral__Bernardo_O_Higgins, Maule, Biobio, La_Araucania, Los_Rios 
    """
    start_date = parse(start_date)
    end_date = parse(end_date)
    
    if start_date < pp.date.min():
        print(f'Please select a start date from {pp.date.min()} onwards')
        return None
    
    elif end_date > pp.date.max():
        print(f'Please select an end date previous to {pp.date.max()}')
        return None
    
    try:
        slice_pp = pp.loc[pp.date.between(start_date, end_date), ['date', region]].sort_values('date')
    except KeyError:
        print(f'Region {region} is not in the region list, please check function documentation to check available regions')
        
        return None
    
    dates = slice_pp.date.values
    reg = slice_pp[region].values
        
    fig, ax = plt.subplots()
    
    ax.plot(dates, reg, label=region)
    fig.autofmt_xdate()
    
    ax.fmt_xdata = mdates.DateFormatter('%Y-%m-%d')

    plt.xlabel('Fecha')
    plt.ylabel('Precipitación')
    plt.title(f'Precipitaciones de la región {region} desde la fecha {start_date.date()} hasta {end_date.date()}')
    plt.legend()
    return plt.show()
```

#### Lags y estadísticas acumuladas
Ahora que hemos definido nuestras variables, podemos crear lags y estadisticas acumuladas de estas variables.

```python
#6 month rolling window for min, max, mean and standard for every column and log
colnames = ['Precio_leche', 'PIB_Agropecuario_silvicola', 'Ocupacion_en_Agricultura_INE', 
            'precipitacion', 'Imacec', 'PIB_no_agro', 'precipitacion_lag']
for column in colnames:
    df_final[column+'__avg'] = df_final[column].rolling(window=6).mean()
    df_final[column+'__std'] = df_final[column].rolling(window=6).std()
    df_final[column+'__min'] = df_final[column].rolling(window=6).min()
    df_final[column+'__max'] = df_final[column].rolling(window=6).max()
    df_final[column+'__log'] = np.log(df_final[column])
    
#12 month difference for and 12 month log difference for 
colnames = ['Precio_leche', 'PIB_Agropecuario_silvicola', 'precipitacion']
for column in colnames:
    df_final[column+'__shift'] = df_final[column] - df_final[column].shift(-12)
    df_final[column+'__log_shift'] = np.log(df_final[column]) - np.log(df_final[column].shift(-12))
```

```python
df_final.replace([np.inf, -np.inf], np.nan, inplace=True)
```

#### 2. Hyperparameter Tunning

Vamos a trabajar para obtener los niveles óptimos de los siguientes parámetros.
- max_depth: La profundidad máxima del arbol, a mayor valor, nos arriesgamos a un mayor overfitting
- min_child_weight: Cantidad mínima de hijos que debe tener una rama para generarse
- subsample: porcentaje de muestreo de las variables
- colsample_bytree: porcentaje de muestreo de las columnas
- eta: learning rate

Obtenido de [blog.cambridgespark.com](https://blog.cambridgespark.com/hyperparameter-tuning-in-xgboost-4ff9100a3b2f)

```python
X = df_model.drop(columns=['milk_shift'])
y = df_model['milk_shift']
X_train = X.loc[:"2016-01-01"]
y_train = y.loc[:"2016-01-01"]
X_test = X.loc["2016-02-01":'2021-01-01']
y_test = y.loc["2016-02-01":'2021-01-01']

y_train.shape, y_test.shape
```

```python
params = {
    # Parameters that we are going to tune.
    'max_depth':6,
    'min_child_weight': 1,
    'eta':.3,
    'subsample': 1,
    'colsample_bytree': 1,
    'objective':'reg:squarederror',
}
```

```python
model.predict(X.iloc[-1, :].values.reshape(1, -1))
```

```python
y_pred = model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)
```

```python
gridsearch_params = [
    (max_depth, min_child_weight, subsample, colsample_bytree, eta)
    for max_depth in range(1,5)
    for min_child_weight in range(1,5)
    for subsample in [i/10. for i in range(1, 7)]
    for colsample_bytree in [i/10. for i in range(7, 11)]
    for eta in [.3, .2, .1, .05, .01, .005]
]
```

```python
min_rmse = float("Inf")
best_params = None
for max_depth, min_child_weight, subsample, colsample_bytree, eta in tqdm(gridsearch_params):
    params['max_depth'] = max_depth
    params['min_child_weight'] = min_child_weight
    params['subsample'] = subsample
    params['colsample_bytree'] = colsample_bytree
    params['eta'] = eta
    # Run CV
    cv_results = xgb.cv(
        params,
        dtrain,
        num_boost_round=999,
        seed=8,
        nfold=5,
        metrics={'rmse'},
        early_stopping_rounds=10
    )
    # Update best MAE
    mean_rmse = cv_results['test-rmse-mean'].min()
    boost_rounds = cv_results['test-rmse-mean'].argmin()
    if mean_rmse < min_rmse:
        min_rmse = mean_rmse
        best_params = (max_depth, min_child_weight, subsample, colsample_bytree, eta)
print(f"""Best params: 
        max_depth:          {best_params[0]}
        min_child_weight:   {best_params[1]}
        subsample:          {best_params[2]}
        colsample_bytree:   {best_params[3]}
        eta:                {best_params[4]}
        Min RMSE:           {min_rmse}""")
```

```python
params['max_depth'], params['min_child_weight'], params['subsample'], params['colsample_bytree'], params['eta'] = best_params
```

```python
model = xgb.train(params, dtrain, num_boost_round=999,evals=[(dtest, "Test")], early_stopping_rounds=99)
```

```python
dval = xgb.DMatrix(X_test, label=y_test)
y_pred = model.predict(dval)
y_pred
```
